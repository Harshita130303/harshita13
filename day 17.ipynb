```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#types of graphs
##1 plot -    line graph,between two parameters(optional )
#2.bar -      bar graph of single parameter
#3.hist -     histogram bunch of data i.e. frequency of single parameter
#4.scatter-   dot graph between two parameters
#5.boxplot/box-   statistical overview in form of box for a single param mean ,outliners,quantiles(25%,75%),IQR(optional-categorical data

#stem-         more like bar graph ,lines instead of bars
# 7.violinplot- violin shaped statistical overview graph
#               spread of data(more like histogra)
#               single parameter (optional categorical -data)  
```

```python
x = np.linspace(0,5,10)
x1 = np.arange(1,11,3)
```

```python
y=x**3
```

```python
plt.plot(x,y)
plt.title('plot')
```

```python
#to run graph in script we have to write plt.show()
```

```python
plt.bar(x,y)
plt.title('bar graph')
```

```python
plt.hist(x,y)
plt.title('histograph')
```

```python
plt.scatter(x,y)
plt.title('scatter')
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.yticks(range(0,200,10))
plt.show()
```

```python
plt.boxplot(y)
plt.title('boxplot')
```

```python
plt.stem(x,y)
plt.title('stem')
```

```python
plt.violinplot(x)
plt.title('violinplot')
```
