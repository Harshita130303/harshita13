```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts().sort_index()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='day',data=tips,hue='sex')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=sum)
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='time',estimator=np.mean)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker')
```
